## Make EDA (Univariate, Bivariate, Multivariate), data clean, missing values management, correlation analysis

```python
# Read the stock prices data using pandas
stock_df = pd.read_csv("../data/stock_data.csv")
```

```python
display(stock_df.head())

stock_df.info()

display(stock_df.describe())

display(stock_df.isnull().sum())
```

```python
df = stock_df.dropna()
```

```python
df['FB'].hist()
plt.title("Histograma de precios de FB")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()


plt.plot(df['Date'], df['FB'], label='Facebook')
plt.plot(df['Date'], df['TWTR'], label='Twitter')
plt.plot(df['Date'], df['NFLX'], label='Netflix')
plt.title("Precios a lo largo del tiempo")
plt.xlabel("Fecha")
plt.ylabel("Precio")
plt.legend()
plt.show()

plt.scatter(df['FB'], df['NFLX'])
plt.title("FB vs NFLX")
plt.xlabel("Facebook")
plt.ylabel("Netflix")
plt.show()

corr = df[['FB', 'TWTR', 'NFLX']].corr()
plt.imshow(corr, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.title("Matriz de correlación")
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.show()
```

# Use seaborn to plot a scatterplot & a countplot

```python
# Seaborn is a visualization library that sits on top of matplotlib
# Seaborn offers enhanced features compared to matplotlib
# https://seaborn.pydata.org/examples/index.html

# import libraries 
import seaborn as sns # Statistical data visualization
```

```python
# Import Cancer data drom the Sklearn library
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer
```

```python
# Create a dataFrame named df_cancer with input/output data
df_cancer = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# Check out the head of the dataframe
df_cancer.head()
```

```python
# Check out the tail of the dataframe
df_cancer.tail()
```

```python
# Plot scatter plot between mean area and mean smoothness
sns.scatterplot(data=df_cancer, x='mean area', y='mean smoothness')
plt.title("Mean Area y Mean Smoothness")
plt.show()
```

```python
# Let's print out countplot to know how many samples belong to class #0 and #1
sns.countplot(data=df_cancer, x='target')
plt.title("Numero de muestras por clase (0 y 1)")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.show()
```

# Using seaborn, plot a scatterplot between the mean readius and mean area
Plese comment the plot analysis

```python
sns.scatterplot(data=df_cancer, x='mean radius', y='mean area')

plt.title("Scatter Plot: Mean Radius vs Mean Area")
plt.xlabel("Mean Radius")
plt.ylabel("Mean Area")
plt.show()
```

# Using seaborn, plot a pairpplot, displot, and a heatmaps/correlations


```python
# Plot the pairplot
sns.pairplot(df_cancer[['mean radius', 'mean texture', 'mean area', 'mean smoothness', 'target']], hue='target')
plt.suptitle("Pairplot de caracteristicas principales", y=1.02)
plt.show()
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter
df_cancer.corr(numeric_only=True)['mean radius'].sort_values(ascending=False)
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.displot(data=df_cancer, x='mean radius', kind='kde', fill=True)
plt.title("Distribucion de Mean Radius (KDE)")
plt.xlabel("Mean Radius")
plt.ylabel("Densidad")
plt.show()

sns.displot(data=df_cancer, x='mean radius', kde=True)
plt.title("Distribucion de Mean Radius con KDE")
plt.xlabel("Mean Radius")
plt.ylabel("Frecuencia / Densidad")
plt.show()
```

```python
#1. Import both datasets using Pandas.
prices_df = pd.read_csv("../data/crypto_daily_prices.csv")
returns_df = pd.read_csv("../data/crypto_daily_returns.csv")
```

```python
#2. Using Matplotlib, plot lineplots that display crypto daily prices on one single figure.
plt.figure(figsize=(10, 6))
for col in ['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price']:
    plt.plot(prices_df['Date'], prices_df[col], label=col)

plt.legend()
plt.xticks(rotation=45)
plt.title("Crypto Daily Prices")
plt.tight_layout()
plt.show()
```

```python
#3. Using Matplotlib, plot crypto daily prices on multiple subplots.
fig, axs = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

for i, col in enumerate(['BTC-USD Price', 'ETH-USD Price', 'LTC-USD Price']):
    axs[i].plot(prices_df['Date'], prices_df[col])
    axs[i].set_title(col)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#4. Using Matplotlib, plot the scatterplot between Bitcoin and Ethereum daily returns.
plt.scatter(returns_df['BTC'], returns_df['ETH'])
plt.xlabel("BTC Returns")
plt.ylabel("ETH Returns")
plt.title("BTC vs ETH Returns")
plt.show()
```

```python
#5. Using Seaborn, plot similar scatterplot between Bitcoin and Ethereum daily returns.
sns.scatterplot(x='BTC', y='ETH', data=returns_df)
plt.title("BTC vs ETH Returns (Seaborn)")
plt.show()
```

```python
#6. Assume that you became bullish on XRP and decided to allocate 60% of your assets in it. 
# You also decided to equally divide the rest of your assets in other coins (BTC, LTC, ADA, and ETH). 
# Using Matplotlib, plot a pie chart that shows these allocations. Use ‘explode’ attribute to increase the separation 
# between BTC and the rest of the portfolio.
values = [60, 10, 10, 10, 10]
labels = ['XRP', 'BTC', 'LTC', 'ADA', 'ETH']
explode = [0, 0.1, 0, 0, 0]

plt.pie(values, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title("Crypto Portfolio Allocation")
plt.show()
```

```python
#7. Using Matplotlib, plot the histogram for ETH returns using 30 bins with red color.
plt.hist(returns_df['ETH'].dropna(), bins=30, color='red')
plt.title("ETH Returns Histogram")
plt.xlabel("Return")
plt.ylabel("Count")
plt.show()
```

```python
#8. Using Seaborn, plot a heatmap that shows the correlations between crypto daily returns.
sns.heatmap(returns_df[['BTC', 'ETH', 'LTC']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
